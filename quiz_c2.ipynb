I can place different audio samples here to test and evaluate the system’s performance across varying inputs.

```python
from datasets import load_dataset

dataset = load_dataset("librispeech_asr", "clean", split="validation")

audio_sample = dataset[0]["audio"]

audio_sample
```

speech → text

```python
from transformers import WhisperProcessor, WhisperForConditionalGeneration
import torch

# load model & processor
processor = WhisperProcessor.from_pretrained("openai/whisper-small")
model = WhisperForConditionalGeneration.from_pretrained("openai/whisper-small")

audio_array = audio_sample["array"]
sampling_rate = audio_sample["sampling_rate"]

inputs = processor(audio_array, sampling_rate=sampling_rate, return_tensors="pt")

with torch.no_grad():
    predicted_ids = model.generate(inputs["input_features"])

transcription = processor.batch_decode(predicted_ids, skip_special_tokens=True)[0]

print("Transcription:")
print(transcription)
```

Baseline

```python
from transformers import T5Tokenizer, T5ForConditionalGeneration
import torch

# load model
tokenizer = T5Tokenizer.from_pretrained("google/flan-t5-base")
model = T5ForConditionalGeneration.from_pretrained("google/flan-t5-base")

# baseline prompt
baseline_prompt = "Summarize the following text:\n" + transcription

inputs = tokenizer(baseline_prompt, return_tensors="pt", truncation=True)

outputs = model.generate(
    inputs["input_ids"],
    max_new_tokens=50
)

baseline_summary = tokenizer.decode(outputs[0], skip_special_tokens=True)

print("Baseline Summary:")
print(baseline_summary)
```

Improved Prompt

```python
improved_prompt = """
You are a professional summarization system.

Rules:
- Use only information from the input text.
- Do NOT add external knowledge.
- Rewrite the sentence in a simpler and more natural form.
- Keep the meaning faithful but do NOT copy the original sentence.
- Output a complete, grammatically correct sentence.

Task:
Summarize the speech below:

""" + transcription

inputs2 = tokenizer(improved_prompt, return_tensors="pt", truncation=True)

outputs2 = model.generate(
    inputs2["input_ids"],
    max_new_tokens=40,
    num_beams=4,
    length_penalty=2.0,
    early_stopping=True
)

improved_summary = tokenizer.decode(outputs2[0], skip_special_tokens=True)

print("Improved Summary:")
print(improved_summary)
```
